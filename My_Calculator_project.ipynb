```python
from math import sqrt
class Calculator:
    
    def __init__(self):
        
        while True:
            
            print("1: Add")
            print("2: Subtract")
            print("3: Multiply")
            print("4: Divide")
            print("5: SquareRoot")
            print("6: Exit")
            print("7: For free mode : free mode is to calculate more than two numbers")

            ch = int(input("Select operation: "))

            #Make sure the user have entered the valid choice
            if ch in (1, 2, 3, 4, 5, 6, 7):

                #first check whether user want to exit
                if(ch == 6):
                    break
                
                #Check if the user want to use the free mode
                if(ch == 7):
                    x = input("Enter the operation you want to calculate: ")
                    print("%s = %s" % (x, self.FreeMode(x)))
                    

                if(ch == 1):
                    a = int(input("Enter first number: "))
                    b = int(input("Enter second number: "))
                    print(a, "+", b, "=", self.add(a, b))
                elif(ch == 2):
                    a = int(input("Enter first number: "))
                    b = int(input("Enter second number: "))
                    print(a, "-", b, "=", self.subtract(a, b))
                elif(ch == 3):
                    a = int(input("Enter first number: "))
                    b = int(input("Enter second number: "))
                    print(a, "*", b, "=", self.multiply(a, b))
                elif(ch == 4):
                    a = int(input("Enter first number: "))
                    b = int(input("Enter second number: "))
                    print(a, "/", b, "=", self.divide(a, b))
                elif(ch == 5):
                    a = int(input("Enter a number: "))
                    print("Square Root of ", a, "=", self.SquareRoot(a))
                    

            else:
                print("Invalid Input")
        
    
    def add(self, a, b):
        return a+b
    
    def subtract(self, a, b):
        return a-b
        
    def multiply(self, a, b):
        return a*b

    def divide(self, a, b):
        return a/b
    
    def SquareRoot(self, a):
        return sqrt(a)
    
    def FreeMode(self, x):
        return eval(x)


Calculator()
```
